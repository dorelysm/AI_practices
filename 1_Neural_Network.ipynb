```python
import pandas as pd

datos = pd.read_csv("celsius.csv")
```

```python
datos.head()
```

```python
X = datos["celsius"].values
y = datos["fahrenheit"].values
```

```python
import tensorflow as tf

modelo = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)), #Entrada -> celsius 1 dato
    tf.keras.layers.Dense(units=1) #Salida -> fahrenheit 1 dato
])

modelo.compile(
    optimizer=tf.keras.optimizers.Adam(.1),
    loss="mean_squared_error"
)
```

```python
entrenamiento = modelo.fit(X, y, epochs=500)
```

```python
import matplotlib.pyplot as plt

plt.xlabel("# Epoca")
plt.ylabel("Magnitud de pérdida")
plt.plot(entrenamiento.history["loss"])
```

```python
import numpy as np
c = 123
prediccion = modelo.predict(np.array([c]))
print(f"Predicción: {c} celsius son {prediccion[0]} fahrenheit")
```
