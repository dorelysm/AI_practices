```python
X = datos["celsius"]
y = datos["fahrenheit"]
```

```python
y
```

```python
X_procesada = X.values.reshape(-1,1)
y_procesada = y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
```

```python
modelo.fit(X_procesada, y_procesada)
```

```python
modelo.predict([[123]])
```
